## 🔄 Panorama General: Condicionales Python vs R

| Concepto | R | Python |
|---|---|---|
| Bloque if | `if (cond) { }` | `if cond:` + indentación |
| else if | `else if (cond) { }` | `elif cond:` |
| else | `else { }` | `else:` |
| Y lógico (escalar) | `&&` | `and` |
| O lógico (escalar) | `\|\|` | `or` |
| Negación | `!` | `not` |
| Y vectorizado | `&` | `&` (solo en NumPy/pandas) |
| O vectorizado | `\|` | `\|` (solo en NumPy/pandas) |
| Pertenencia | `x %in% v` | `x in lista` |
| No pertenencia | `!(x %in% v)` | `x not in lista` |
| Ternario | `ifelse(cond, a, b)` | `a if cond else b` |
| Nulo/vacío | `is.null(x)` | `x is None` |
| Vectorial condicional | `ifelse(v > 0, 'pos', 'neg')` | `['pos' if x>0 else 'neg' for x in v]` |

::: {.callout-warning title="⚠️ La diferencia más importante"}
En R existen **dos versiones** de cada operador lógico:
- `&` y `|` son **vectorizados** (operan sobre vectores completos)
- `&&` y `||` son **escalares** (solo evalúan el primer elemento)

En Python, `and`, `or`, `not` son **siempre escalares**. Para operaciones vectorizadas con NumPy/pandas se usan `&`, `|`, `~`, pero eso es para capítulos posteriores.

Usar `&` y `|` en una condición de `if` de Python puede dar resultados inesperados con valores no booleanos.
:::

## 5.1 Condicionales Básicos

### 5.1.1 Estructura `if` / `elif` / `else`

````python
# Estructura básica
# Python usa dos puntos : y sangría (4 espacios) en lugar de llaves {}

edad = 22

if edad < 13:
    print("Niño/a")
elif edad < 18:
    print("Adolescente")
elif edad < 65:
    print("Adulto/a")         # ← se ejecuta este bloque
else:
    print("Adulto/a mayor")

print("Esta línea siempre se ejecuta — está fuera del if")
````

```r
  cat("Niño/a\n")
} else if (edad < 18) {    # 'else if' (dos palabras) en R
  cat("Adolescente\n")
} else if (edad < 65) {
  cat("Adulto/a\n")
} else {
  cat("Adulto/a mayor\n")
}
```

```python
# En Python — indentación define los bloques
edad = 22

if edad < 13:              # sin paréntesis (opcionales pero innecesarios)
    print("Niño/a")
elif edad < 18:             # 'elif' (una sola palabra) en Python
    print("Adolescente")
elif edad < 65:
    print("Adulto/a")
else:
    print("Adulto/a mayor")
```
Diferencias clave: sin paréntesis en la condición, `elif` en lugar de `else if`, dos puntos al final de cada rama, y la **indentación como parte de la sintaxis**.

### 5.1.2 Operadores de comparación

````python
# Los operadores de comparación retornan True o False
x = 10
y = 7

print(f"x = {x}, y = {y}")
print(f"x == y  → {x == y}")    # igualdad (== no =)
print(f"x != y  → {x != y}")    # diferencia
print(f"x >  y  → {x > y}")     # mayor que
print(f"x <  y  → {x < y}")     # menor que
print(f"x >= y  → {x >= y}")    # mayor o igual
print(f"x <= y  → {x <= y}")    # menor o igual

# Encadenamiento de comparaciones — ¡Python lo soporta nativamente!
nota = 75
print(f"\n0 <= nota <= 100: {0 <= nota <= 100}")   # ✅ pythónico
print(f"60 < nota < 80:   {60 < nota < 80}")

# En R esto no funciona directamente:
# 0 <= nota & nota <= 100   # así sería en R
````

::: {.callout-note title="R vs Python — Comparaciones encadenadas"}

```python
# En Python — el encadenamiento es nativo y muy legible
nota = 75
0 <= nota <= 100           # ✅ equivalente matemático directo
60 < nota < 80             # ✅ también válido
```
El encadenamiento de comparaciones de Python es una ventaja real sobre R para rangos numéricos.
:::

### 5.1.3 Comparaciones con strings

````python
# Las comparaciones de strings son sensibles a mayúsculas (como en R)
lenguaje = "Python"

print(lenguaje == "Python")           # True
print(lenguaje == "python")           # False — case-sensitive
print(lenguaje == "PYTHON")           # False

# La solución: normalizar antes de comparar
entrada_usuario = "PYTHON"
if entrada_usuario.lower() == "python":
    print("¡Correcto!")

# Comparaciones lexicográficas (como en R)
print("\n--- Comparaciones lexicográficas ---")
print("banana" < "manzana")     # True (b antes que m)
print("abc" < "abd")            # True (c antes que d)
print("Z" < "a")               # True (Z=90, a=97 en ASCII)

# Verificar si un string contiene otro (operador in)
frase = "Python es poderoso y elegante"
print("\n--- Pertenencia en strings ---")
print("Python" in frase)         # True
print("Java" in frase)           # False
print("elegante" in frase)       # True
````

## 5.2 Operadores Lógicos

### 5.2.1 `and`, `or`, `not`

````python
# Los operadores lógicos en Python son palabras: and, or, not
# Son equivalentes a &&, ||, ! en R (versión escalar)

temperatura = 25
humedad     = 75
dia_laboral = True

# and — ambas condiciones deben ser True
if temperatura > 20 and humedad < 60:
    print("Clima perfecto para salir")
elif temperatura > 20 and humedad >= 60:
    print("Caluroso y húmedo")   # ← se ejecuta este
else:
    print("Mejor quedarse en casa")

# or — al menos una condición debe ser True
dia = "sábado"
if dia == "sábado" or dia == "domingo":
    print("¡Es fin de semana!")

# not — niega la condición
llueve = False
if not llueve:
    print("No está lloviendo — buen momento para salir")

# Combinaciones complejas — usar paréntesis para claridad
edades_validas = True
datos_completos = True
permisos = False

if (edades_validas and datos_completos) and not permisos:
    print("Datos ok pero faltan permisos")
````

````python
# Evaluación de cortocircuito (short-circuit evaluation)
# Python deja de evaluar en cuanto conoce el resultado

# Con 'and': si el primero es False, no evalúa el segundo
# Con 'or':  si el primero es True, no evalúa el segundo

# Esto es útil para evitar errores:
lista = []

# Sin cortocircuito: lista[0] daría IndexError si lista está vacía
# Con and: si 'lista' es falsy, la segunda condición no se evalúa
if lista and lista[0] > 5:
    print("Primer elemento mayor que 5")
else:
    print("Lista vacía o primer elemento no cumple")

# Patrón común: valor por defecto con or
nombre_usuario = ""     # podría venir de input del usuario
nombre = nombre_usuario or "Invitado"   # si nombre_usuario es falsy, usa "Invitado"
print(f"Bienvenido, {nombre}")

nombre_usuario = "Ana"
nombre = nombre_usuario or "Invitado"
print(f"Bienvenido, {nombre}")
````

::: {.callout-warning title="⚠️ R vs Python — Operadores lógicos: la diferencia más importante"}
Esta es una fuente frecuente de bugs al migrar de R a Python:

```r
# En R hay DOS conjuntos de operadores lógicos:
# Escalares (para if/while): &&, ||, !
x <- 5
if (x > 3 && x < 10) cat("en rango\n")  # ✅ correcto para escalares

# Vectorizados (para vectores completos): &, |, !
v <- c(1, 5, 3, 8, 2)
v[v > 3 & v < 7]    # [5, 3] — filtra el vector ✅
```
```python
# En Python los operadores lógicos para if/while son palabras:
x = 5
if x > 3 and x < 10:    # ✅ correcto
    print("en rango")

# & y | en Python son operadores de BITS — pueden dar resultados inesperados
# con enteros que no sean 0/1:
if x > 3 & x < 10:      # ⚠️ & tiene mayor precedencia que > y <
    pass                  # se evalúa como x > (3 & x) < 10 — ¡no lo esperado!

# Para listas/vectores de Python puro:
v = [1, 5, 3, 8, 2]
filtrado = [n for n in v if n > 3 and n < 7]   # ✅ usar and aquí

# Para NumPy/pandas (capítulo posterior): ahí sí se usa & y |
```

| Operación | R correcto | Python correcto |
|---|---|---|
| `if` escalar | `x > 3 && x < 10` | `x > 3 and x < 10` |
| Filtrar vector | `v[v > 3 & v < 7]` | `[n for n in v if n > 3 and n < 7]` |
| Negación escalar | `!es_valido` | `not es_valido` |

## 5.3 Valores Truthy y Falsy

### 5.3.1 ¿Qué evalúa como False en Python?

````python
# En Python, estos valores se comportan como False en un contexto booleano:
# False, None, 0, 0.0, 0j, "", [], {}, (), set()

valores_falsy = [False, None, 0, 0.0, "", [], {}, ()]
valores_truthy = [True, 1, -1, 3.14, "hola", [0], {"a": 1}, (1,)]

print("--- Valores FALSY (se comportan como False) ---")
for v in valores_falsy:
    resultado = "False" if not v else "True"
    print(f"  {repr(v):>10}  →  {resultado}")

print("\n--- Valores TRUTHY (se comportan como True) ---")
for v in valores_truthy:
    resultado = "True" if v else "False"
    print(f"  {repr(v):>10}  →  {resultado}")
````

````python
# Uso idiomático de Truthy/Falsy — hace el código más conciso

# Verificar si una lista tiene elementos
resultados = [85, 92, 78]

# ❌ Forma verbosa (al estilo R)
if len(resultados) > 0:
    print("Hay resultados")

# ✅ Forma pythónica
if resultados:
    print("Hay resultados")

# Verificar strings vacíos
nombre = ""
if not nombre:   # más limpio que: if nombre == ""
    print("El nombre está vacío")

# Verificar None
valor = None
if valor is None:         # ✅ correcto para None
    print("No hay valor")
# if not valor también funciona, pero 'is None' es más explícito y seguro
# (evita confundir None con 0 o "")

# Verificar si una variable tiene valor útil
def buscar_usuario(id):
    if id == 1:
        return "Ana García"
    return None   # no encontrado

usuario = buscar_usuario(99)
if usuario:   # None es falsy
    print(f"Usuario encontrado: {usuario}")
else:
    print("Usuario no encontrado")
````

::: {.callout-note title="R vs Python — Valores Falsy"}
```r
# En R, los valores "falsos" en un if son:
# FALSE, NA (con advertencia), NULL (con error), 0, 0L, 0.0

# En R, una lista vacía en if() da ERROR
if (list()) {}   # Error: argument is of length zero

# En R, NULL en if() da ERROR
if (NULL) {}     # Error: argument is of length zero

# La forma idiomática en R para verificar vacío:
if (length(v) > 0) {}       # vector
if (!is.null(x)) {}          # NULL
if (!is.na(x)) {}            # NA
if (nchar(s) > 0) {}         # string
```
```python
# En Python, estas verificaciones son más concisas:
if v:           # lista/vector vacío → False automáticamente
if x is not None:  # None
if s:           # string vacío → False automáticamente
```
Python es más permisivo aquí: cualquier colección vacía evaluará como `False` directamente en un `if`.

## 5.4 Operador Ternario

### 5.4.1 Expresión condicional en una línea

````python
# Sintaxis: valor_si_true if condicion else valor_si_false
# Lee como: "dame X si condición, si no dame Y"

edad = 20
estado = "mayor de edad" if edad >= 18 else "menor de edad"
print(f"{edad} años → {estado}")

# Más ejemplos
nota = 75
aprobado = "✅ Aprobado" if nota >= 70 else "❌ Reprobado"
print(f"Nota {nota}: {aprobado}")

precio  = 450
descuento = precio * 0.9 if precio > 300 else precio
print(f"Precio ${precio} → con descuento: ${descuento}")

# Se puede usar dentro de f-strings
for n in [3, 7, 12, 15, 21]:
    tipo = "par" if n % 2 == 0 else "impar"
    print(f"{n:3d} es {tipo}")
````

::: {.callout-note title="R vs Python — Operador ternario"}
```r
# En R — ifelse() vectorizado (trabaja sobre vectores completos)
nota <- 75
aprobado <- ifelse(nota >= 70, "Aprobado", "Reprobado")  # escalar también funciona

# Para vectores:
notas   <- c(65, 85, 72, 55, 91)
estados <- ifelse(notas >= 70, "Aprobado", "Reprobado")  # vectorizado ✨
```
```python
# En Python — operador ternario (escalar)
nota = 75
aprobado = "Aprobado" if nota >= 70 else "Reprobado"

# Para listas completas — list comprehension:
notas   = [65, 85, 72, 55, 91]
estados = ["Aprobado" if n >= 70 else "Reprobado" for n in notas]
```

| Operación | R | Python |
|---|---|---|
| Escalar | `ifelse(cond, a, b)` | `a if cond else b` |
| Vectorial | `ifelse(v > 0, "pos", "neg")` | `["pos" if x>0 else "neg" for x in v]` |

El `ifelse()` de R es automáticamente vectorizado — una ventaja real de R para este caso.

## 5.5 Condicionales con Listas

### 5.5.1 Operadores `in` y `not in`

````python
# in y not in verifican pertenencia en cualquier colección
lenguajes_web  = ['javascript', 'python', 'php', 'ruby', 'typescript']
lenguajes_data = ['python', 'r', 'julia', 'sql', 'scala']

lenguaje = 'python'

print(f"'{lenguaje}' in web:  {lenguaje in lenguajes_web}")
print(f"'{lenguaje}' in data: {lenguaje in lenguajes_data}")
print(f"'java' not in web:    {'java' not in lenguajes_web}")

# Uso práctico: validar entradas
def verificar_lenguaje(lang):
    web  = ['javascript', 'python', 'php', 'typescript']
    data = ['python', 'r', 'julia', 'sql']

    lang = lang.lower()
    en_web  = lang in web
    en_data = lang in data

    if en_web and en_data:
        return f"'{lang}' es de uso general (web + datos)"
    elif en_web:
        return f"'{lang}' es principalmente para web"
    elif en_data:
        return f"'{lang}' es principalmente para datos"
    else:
        return f"'{lang}' no está en las listas conocidas"

for l in ['Python', 'JavaScript', 'R', 'Go', 'SQL']:
    print(verificar_lenguaje(l))
````

::: {.callout-note title="R vs Python — Operador de pertenencia"}
```r
# En R — %in% (operador de infijo)
lenguajes_web <- c('javascript', 'python', 'php')
'python' %in% lenguajes_web      # TRUE
!('java' %in% lenguajes_web)     # TRUE

# Para vectores completos (vectorizado):
v <- c('python', 'java', 'r')
v %in% lenguajes_web             # c(TRUE, FALSE, FALSE)
```
```python
# En Python — operadores in / not in
lenguajes_web = ['javascript', 'python', 'php']
'python' in lenguajes_web        # True
'java' not in lenguajes_web      # True

# Para listas completas:

[x in lenguajes_web for x in v]  # [True, False, False]
```
`%in%` de R y `in` de Python son equivalentes directos en funcionalidad.

### 5.5.2 `any()` y `all()` — condiciones sobre colecciones completas

````python
# any() — True si AL MENOS UN elemento cumple la condición
# all() — True si TODOS los elementos cumplen la condición
# Son equivalentes a any() y all() de R

calificaciones = [72, 85, 91, 68, 74, 88, 95, 63]

# ¿Alguna calificación perfecta (100)?
alguna_perfecta = any(c == 100 for c in calificaciones)
print(f"¿Alguna calificación de 100? {alguna_perfecta}")

# ¿Alguna reprobatoria (< 70)?
alguna_reprobatoria = any(c < 70 for c in calificaciones)
print(f"¿Alguna reprobatoria?        {alguna_reprobatoria}")

# ¿Todos aprobaron?
todos_aprobaron = all(c >= 70 for c in calificaciones)
print(f"¿Todos aprobaron?            {todos_aprobaron}")

# ¿Todos están en el rango válido 0-100?
todos_validos = all(0 <= c <= 100 for c in calificaciones)
print(f"¿Todos en rango 0-100?       {todos_validos}")

# Combinando con if
print("\n--- En condicionales ---")
if any(c < 60 for c in calificaciones):
    print("⚠️  Hay calificaciones muy bajas")

if not all(c >= 70 for c in calificaciones):
    reprobados = [c for c in calificaciones if c < 70]
    print(f"❌ Hay {len(reprobados)} reprobados: {reprobados}")
````

::: {.callout-note title="R vs Python — any() y all()"}
```r
# En R — any() y all() trabajan sobre vectores directamente
califs <- c(72, 85, 91, 68, 74)
any(califs < 70)           # TRUE — vectorización directa
all(califs >= 70)          # FALSE
all(califs >= 0 & califs <= 100)  # TRUE
```
```python
# En Python — se usa con generadores (expresiones entre paréntesis)
califs = [72, 85, 91, 68, 74]
any(c < 70 for c in califs)        # True — generador
all(c >= 70 for c in califs)       # False
all(0 <= c <= 100 for c in califs) # True
```
La funcionalidad es idéntica. La diferencia es que en R la vectorización es automática, mientras que en Python necesitas el generador `for c in califs`.

## 5.6 Condicionales Anidados

### 5.6.1 Cuándo anidar y cuándo no

````python
# Los condicionales anidados funcionan, pero hay que usarlos con moderación
# La regla de oro: máximo 2-3 niveles de anidamiento

puntaje = 85
es_miembro = True
anios_membresia = 3

# --- Forma anidada (legible si no es muy profunda) ---
if puntaje >= 70:
    if es_miembro:
        if anios_membresia >= 2:
            descuento = 20
        else:
            descuento = 10
    else:
        descuento = 5
else:
    descuento = 0

print(f"Descuento: {descuento}%")
````

````python
# --- Forma aplanada (generalmente más clara) ---
# Usa 'and' para combinar condiciones en lugar de anidar

puntaje = 85
es_miembro = True
anios_membresia = 3

if puntaje >= 70 and es_miembro and anios_membresia >= 2:
    descuento = 20
elif puntaje >= 70 and es_miembro:
    descuento = 10
elif puntaje >= 70:
    descuento = 5
else:
    descuento = 0

print(f"Descuento: {descuento}%")

# Principio de "guardas" (early return) — muy pythónico en funciones:
def calcular_descuento(puntaje, es_miembro, anios):
    """Calcula el descuento aplicable."""
    if puntaje < 70:
        return 0          # guarda: salida temprana
    if not es_miembro:
        return 5          # guarda: salida temprana
    if anios >= 2:
        return 20
    return 10             # caso base

print(f"Descuento (función): {calcular_descuento(85, True, 3)}%")
print(f"Descuento (función): {calcular_descuento(60, True, 3)}%")
````

## 5.7 Condicionales en Loops y Comprehensions

````python
# Condicionales dentro de loops for
toppings_disponibles = ['pepperoni', 'champiñones', 'aceitunas', 'queso extra', 'jalapeños']
pedido = ['pepperoni', 'anchoas', 'champiñones', 'piña']

print("=== VERIFICACIÓN DE PEDIDO ===")
disponibles_en_pedido  = []
no_disponibles_en_pedido = []

for topping in pedido:
    if topping in toppings_disponibles:
        disponibles_en_pedido.append(topping)
        print(f"  ✅ {topping.title()}: disponible")
    else:
        no_disponibles_en_pedido.append(topping)
        print(f"  ❌ {topping.title()}: no disponible")

if no_disponibles_en_pedido:
    print(f"\n⚠️  No podemos agregar: {', '.join(no_disponibles_en_pedido)}")
else:
    print("\n✅ Todos los ingredientes están disponibles")
````

````python
# Condicionales en list comprehensions (repaso del capítulo 4)
numeros = list(range(-10, 11))

# Filtrar (if al final)
positivos = [n for n in numeros if n > 0]
print(f"Positivos:     {positivos}")

# Transformar condicionalmente (if-else en la expresión)
signos = ["pos" if n > 0 else ("neg" if n < 0 else "cero") for n in numeros]
print(f"Signos:        {signos}")

# Ejemplo práctico: clasificar calificaciones
califs = [92, 65, 78, 88, 55, 72, 95, 61, 84, 70]

letras = [
    'A' if c >= 90 else
    'B' if c >= 80 else
    'C' if c >= 70 else
    'D' if c >= 60 else 'F'
    for c in califs
]
print(f"\nCalificaciones: {califs}")
print(f"Letras:         {letras}")
````

## 5.8 Errores Comunes

````python
# Error 1: Usar = en lugar de == para comparar
x = 10

# ❌ Error de sintaxis (Python no lo permite en if)
# if x = 10:   # SyntaxError

# ✅ Correcto
if x == 10:
    print("x es 10")

# En R, el linter te avisa de esto pero no siempre previene el bug silencioso:
# if (x = 10) {}  # Error en R también, pero el mensaje puede ser confuso
````

````python
# Error 2: Usar & y | en lugar de and y or en un if
x = 5

# ⚠️ Peligroso — & tiene mayor precedencia que > y <
# if x > 3 & x < 10:   # se evalúa como x > (3 & 10) < 10
#                       # (3 & 10) = 2 en binario → x > 2 < 10 → True siempre

# ✅ Correcto
if x > 3 and x < 10:
    print("En rango")

# ✅ También correcto (encadenamiento)
if 3 < x < 10:
    print("En rango (encadenado)")
````

````python
# Error 5: Comparar con True/False explícitamente (antipatrón)
esta_activo = True

# ❌ Redundante
if esta_activo == True:
    print("Activo")

# ✅ Pythónico
if esta_activo:
    print("Activo")

# ❌ Redundante
if esta_activo == False:
    print("Inactivo")

# ✅ Pythónico
if not esta_activo:
    print("Inactivo")

# La excepción: cuando necesitas distinguir entre False y None
resultado = None   # podría ser None, True o False

if resultado is False:    # solo cuando resultado es exactamente False
    print("Explícitamente falso")
elif resultado is None:
    print("Sin resultado aún")
````

```python
def validar_registro(nombre, email, edad, contrasena, pais):
    errores = []

    # Nombre
    if not nombre:
        errores.append("Nombre vacío")
    elif len(nombre) < 3:
        errores.append(f"Nombre muy corto ({len(nombre)} chars, mínimo 3)")

    if len(contrasena) < 8:

        errores.append("Contraseña debe tener al menos una mayúscula")

    if pais not in paises_permitidos:
        errores.append(f"País '{pais}' no está en la lista de países permitidos")
```

## 🚀 Mini Proyecto: Calculadora de IMC con Reporte Completo

````python
# Mini Proyecto: Análisis nutricional con condicionales
# Combina todo lo aprendido: if/elif/else, and/or, any/all,
# operador ternario, in/not in y comprehensions condicionales

def clasificar_imc(imc):
    """Retorna (clasificación, emoji, consejo) según el IMC."""
    if imc < 16:
        return "Desnutrición severa", "🔴", "Atención médica urgente"
    elif imc < 18.5:
        return "Bajo peso", "🟡", "Aumentar ingesta calórica con supervisión"
    elif imc < 25:
        return "Peso normal", "🟢", "¡Mantén tus hábitos saludables!"
    elif imc < 30:
        return "Sobrepeso", "🟡", "Aumentar actividad física gradualmente"
    elif imc < 35:
        return "Obesidad grado I", "🟠", "Consulta a nutriólogo"
    elif imc < 40:
        return "Obesidad grado II", "🔴", "Atención médica recomendada"
    else:
        return "Obesidad grado III", "🔴", "Atención médica urgente"

def calcular_y_reportar(nombre, peso_kg, altura_m):
    """Calcula IMC y genera reporte."""
    if peso_kg <= 0 or altura_m <= 0:
        return f"❌ {nombre}: datos inválidos"
    if not (1.0 <= altura_m <= 2.5):
        return f"❌ {nombre}: altura fuera de rango realista"

    imc = peso_kg / (altura_m ** 2)
    clasif, emoji, consejo = clasificar_imc(imc)

    # Rango de peso saludable para esta altura
    peso_min = round(18.5 * altura_m ** 2, 1)
    peso_max = round(24.9 * altura_m ** 2, 1)
    en_rango = peso_min <= peso_kg <= peso_max

    resultado = [
        f"{emoji} {nombre}",
        f"   Peso: {peso_kg} kg | Altura: {altura_m} m | IMC: {imc:.1f}",
        f"   Clasificación: {clasif}",
        f"   Rango saludable: {peso_min}–{peso_max} kg {'✅' if en_rango else '⚠️'}",
        f"   Consejo: {consejo}",
    ]
    return "\n".join(resultado)


personas = [
    ("Ana",     55, 1.65),
    ("Carlos",  95, 1.78),
    ("Beatriz", 48, 1.72),
    ("Diego",  130, 1.70),
    ("Elena",   58, 1.60),
]

print("=" * 55)
print(f"{'REPORTE DE ÍNDICE DE MASA CORPORAL':^55}")
print("=" * 55)

for datos in personas:
    print()
    print(calcular_y_reportar(*datos))

# Resumen grupal usando condicionales y comprehensions
imcs = [datos[1] / datos[2]**2 for datos in personas]
print("\n" + "=" * 55)
print("RESUMEN DEL GRUPO")
print("-" * 55)
print(f"Promedio IMC:   {sum(imcs)/len(imcs):.1f}")
print(f"En peso normal: {sum(18.5 <= imc < 25 for imc in imcs)} personas")
print(f"Con sobrepeso+: {sum(imc >= 25 for imc in imcs)} personas")
print(f"¿Algún caso urgente (IMC < 16 o >= 40)? {any(imc < 16 or imc >= 40 for imc in imcs)}")
````

## 🗒️ Resumen del Capítulo

| Concepto | R | Python |
|---|---|---|
| Bloque if | `if (cond) { }` | `if cond:` + sangría |
| else if | `else if (cond) { }` | `elif cond:` |
| Y lógico (if) | `&&` | `and` |
| O lógico (if) | `\|\|` | `or` |
| Negación | `!` | `not` |
| Pertenencia | `x %in% v` | `x in v` |
| No pertenencia | `!(x %in% v)` | `x not in v` |
| Ternario escalar | `ifelse(c, a, b)` | `a if c else b` |
| Ternario vectorial | `ifelse(v > 0, a, b)` | `[a if x>0 else b for x in v]` |
| Alguno cumple | `any(v > 0)` | `any(x > 0 for x in v)` |
| Todos cumplen | `all(v > 0)` | `all(x > 0 for x in v)` |
| Verificar vacío | `length(v) == 0` | `not v` o `len(v) == 0` |
| Verificar None | `is.null(x)` | `x is None` |
| Encadenamiento | `x >= 0 & x <= 10` | `0 <= x <= 10` ✨ |
