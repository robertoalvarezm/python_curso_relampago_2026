## 7.1 Definición Básica de Funciones

```python
# Sintaxis básica de una función
# def nombre_funcion(parametros):
#     "docstring"  ← documentación
#     código
#     return valor  ← opcional

def saludar():
    """Muestra un saludo simple."""
    print("¡Hola desde la función!")

# Llamar la función
saludar()
saludar()  # Se puede llamar múltiples veces

# Ver la documentación
help(saludar)
print(saludar.__doc__)
```

```python
# Función con parámetros
def saludar_persona(nombre, lenguaje="Python"):
    """Saluda a una persona mencionando su lenguaje favorito.
    
    Args:
        nombre: El nombre de la persona
        lenguaje: Lenguaje favorito (por defecto: Python)
    """
    print(f"Hola, {nombre}! Veo que programas en {lenguaje}.")

# Diferentes formas de llamar la función
saludar_persona("Ana")                      # Usa valor por defecto
saludar_persona("Carlos", "Java")           # Argumento posicional
saludar_persona("Beatriz", lenguaje="Go")   # Argumento keyword
saludar_persona(lenguaje="Rust", nombre="Diego")  # Orden flexible con keywords
```

## 7.2 Retorno de Valores

```python
# Funciones que retornan valores
def calcular_area_rectangulo(ancho, alto):
    """Calcula el área de un rectángulo."""
    return ancho * alto

# El valor retornado puede usarse directamente
area = calcular_area_rectangulo(10, 5)
print(f"Área: {area} m²")

# O en expresiones
if calcular_area_rectangulo(3, 4) > 10:
    print("El rectángulo es grande")

# Retorno múltiple (usando tupla)
def estadisticas(numeros):
    """Retorna estadísticas básicas de una lista."""
    return min(numeros), max(numeros), sum(numeros) / len(numeros)

datos = [15, 8, 23, 42, 4, 16]
minimo, maximo, promedio = estadisticas(datos)
print(f"Min: {minimo}, Max: {maximo}, Promedio: {promedio:.1f}")
```

```python
# Retornar diccionarios — muy útil para retornar datos estructurados
def crear_perfil(nombre, edad, **kwargs):
    """Crea un perfil de usuario.
    
    Args:
        nombre: Nombre del usuario
        edad: Edad del usuario
        **kwargs: Atributos adicionales opcionales
    """
    perfil = {
        'nombre': nombre,
        'edad': edad,
    }
    # **kwargs permite pasar atributos adicionales
    for clave, valor in kwargs.items():
        perfil[clave] = valor
    return perfil

ana = crear_perfil("Ana García", 28, ciudad="Madrid", lenguaje="Python")
carlos = crear_perfil("Carlos", 34, empresa="TechCorp")

print(ana)
print(carlos)
```

## 7.3 Argumentos Especiales

```python
# *args — número variable de argumentos posicionales
def sumar(*numeros):
    """Suma cualquier cantidad de números."""
    print(f"Números recibidos: {numeros}")  # Es una tupla
    return sum(numeros)

print(sumar(1, 2))
print(sumar(1, 2, 3, 4, 5))
print(sumar(10, 20, 30, 40, 50, 60))

# **kwargs — número variable de argumentos keyword
def mostrar_info(**datos):
    """Muestra información de forma dinámica."""
    for clave, valor in datos.items():
        print(f"  {clave}: {valor}")

mostrar_info(nombre="Ana", edad=28, ciudad="Madrid")
mostrar_info(producto="Laptop", precio=15999, stock=10)
```

## 7.4 Funciones como Objetos y Funciones Lambda

```python
# Las funciones son objetos — pueden asignarse y pasarse
def doblar(x):
    return x * 2

def triplicar(x):
    return x * 3

operaciones = [doblar, triplicar]  # Lista de funciones!

for op in operaciones:
    print(f"{op.__name__}(5) = {op(5)}")

# Funciones como argumento (callbacks)
numeros = [3, 1, 4, 1, 5, 9, 2, 6]
print(sorted(numeros))                          # Orden natural
print(sorted(numeros, key=lambda x: -x))        # Orden descendente

# Lambda — función anónima de una línea
# lambda argumentos: expresión
cuadrado = lambda x: x ** 2
print(cuadrado(7))

# Muy útil con sorted, map, filter
nombres = ['Ana', 'carlos', 'Beatriz', 'diego']
print(sorted(nombres, key=lambda n: n.lower()))
```

### Ejercicio 7.1: Calculadora Modular

```python
# Ejercicio 7.1 — Calculadora con funciones
def suma(a, b): return a + b
def resta(a, b): return a - b
def multiplicacion(a, b): return a * b

def division(a, b):
    """División con manejo de error."""
    if b == 0:
        return "Error: División por cero"
    return a / b

def potencia(base, exponente=2):
    """Potencia con exponente por defecto 2."""
    return base ** exponente

def raiz_cuadrada(n):
    """Raíz cuadrada (solo números no negativos)."""
    if n < 0:
        return "Error: Número negativo"
    return n ** 0.5

# Tabla de operaciones
print(f"{'Operación':<25} {'Resultado':>10}")
print("-" * 37)
operaciones = [
    ("Suma 15 + 8", suma(15, 8)),
    ("Resta 20 - 7", resta(20, 7)),
    ("Multiplicación 6 × 9", multiplicacion(6, 9)),
    ("División 100 ÷ 4", division(100, 4)),
    ("División 5 ÷ 0", division(5, 0)),
    ("Potencia 3²", potencia(3)),
    ("Potencia 2¹⁰", potencia(2, 10)),
    ("√144", raiz_cuadrada(144)),
]

for nombre, resultado in operaciones:
    print(f"{nombre:<25} {str(resultado):>10}")
```

### Ejercicio 7.2: Procesador de Texto

```python
# Ejercicio 7.2 — Funciones para procesamiento de texto
def contar_palabras(texto):
    """Cuenta palabras en el texto."""
    return len(texto.split())

def contar_vocales(texto):
    """Cuenta vocales (incluyendo acentuadas)."""
    vocales = 'aeiouáéíóúAEIOUÁÉÍÓÚ'
    return sum(1 for c in texto if c in vocales)

def es_palindromo(texto):
    """Verifica si el texto es palíndromo (ignora espacios y mayúsculas)."""
    limpio = ''.join(c.lower() for c in texto if c.isalpha())
    return limpio == limpio[::-1]

def invertir_palabras(texto):
    """Invierte el orden de las palabras."""
    return ' '.join(reversed(texto.split()))

def capitalizar_palabras(texto):
    """Capitaliza solo la primera letra de cada palabra."""
    return ' '.join(p.capitalize() for p in texto.split())

# Pruebas
textos = [
    "Anita lava la tina",
    "Python es un lenguaje increíble",
    "A man a plan a canal Panama",
]

for texto in textos:
    print(f"
📝 Texto: '{texto}'")
    print(f"   Palabras: {contar_palabras(texto)}")
    print(f"   Vocales: {contar_vocales(texto)}")
    print(f"   ¿Palíndromo? {'Sí' if es_palindromo(texto) else 'No'}")
    print(f"   Invertido: '{invertir_palabras(texto)}'")
```

## 🚀 Mini Proyecto: Sistema de Gestión de Biblioteca

```python
# MINI PROYECTO: Sistema de Biblioteca con Funciones
biblioteca = {}

def agregar_libro(titulo, autor, anio, copias=1):
    """Agrega un libro a la biblioteca."""
    clave = titulo.lower()
    if clave in biblioteca:
        biblioteca[clave]['copias'] += copias
        print(f"📚 Copias adicionales de '{titulo}' agregadas. Total: {biblioteca[clave]['copias']}")
    else:
        biblioteca[clave] = {
            'titulo': titulo,
            'autor': autor,
            'anio': anio,
            'copias': copias,
            'prestados': 0
        }
        print(f"✅ '{titulo}' de {autor} agregado a la biblioteca")

def prestar_libro(titulo):
    """Presta un libro si hay copias disponibles."""
    clave = titulo.lower()
    if clave not in biblioteca:
        return f"❌ '{titulo}' no existe en la biblioteca"
    libro = biblioteca[clave]
    disponibles = libro['copias'] - libro['prestados']
    if disponibles > 0:
        libro['prestados'] += 1
        return f"✅ '{libro['titulo']}' prestado. Quedan {disponibles - 1} copias"
    else:
        return f"⚠️  '{libro['titulo']}' no tiene copias disponibles"

def devolver_libro(titulo):
    """Registra la devolución de un libro."""
    clave = titulo.lower()
    if clave in biblioteca and biblioteca[clave]['prestados'] > 0:
        biblioteca[clave]['prestados'] -= 1
        return f"✅ '{biblioteca[clave]['titulo']}' devuelto correctamente"
    return f"❌ Error al devolver '{titulo}'"

def mostrar_catalogo():
    """Muestra el catálogo completo."""
    print(f"
{'='*65}")
    print(f"{'CATÁLOGO DE BIBLIOTECA':^65}")
    print(f"{'='*65}")
    print(f"{'Título':<30} {'Autor':<20} {'Disp.':>5} {'Prest.':>6}")
    print(f"{'-'*65}")
    for libro in sorted(biblioteca.values(), key=lambda x: x['titulo']):
        disponibles = libro['copias'] - libro['prestados']
        print(f"{libro['titulo']:<30} {libro['autor']:<20} {disponibles:>5} {libro['prestados']:>6}")
    print(f"{'='*65}")
    print(f"Total de títulos: {len(biblioteca)}")

# Demo del sistema
agregar_libro("El Principito", "Antoine de Saint-Exupéry", 1943, 3)
agregar_libro("Python Crash Course", "Eric Matthes", 2023, 2)
agregar_libro("Clean Code", "Robert C. Martin", 2008, 1)
agregar_libro("El Principito", "Antoine de Saint-Exupéry", 1943, 2)

print("
--- PRÉSTAMOS ---")
print(prestar_libro("El Principito"))
print(prestar_libro("Clean Code"))
print(prestar_libro("Clean Code"))

print("
--- DEVOLUCIÓN ---")
print(devolver_libro("Clean Code"))

mostrar_catalogo()
```
