## 4.1 Loops con Listas

```python
# for loop básico — itera sobre cada elemento
lenguajes = ['Python', 'Java', 'C++', 'Go', 'Rust']

print("Lenguajes que conozco:")
for lenguaje in lenguajes:
    print(f"  → {lenguaje}")

print(f"
Total: {len(lenguajes)} lenguajes")
```

```python
# enumerate() — obtener índice y valor simultáneamente
# Similar a: for (int i = 0; i < arr.length; i++) en Java, pero más elegante
frutas = ['manzana', 'banana', 'cereza']

for indice, fruta in enumerate(frutas):
    print(f"{indice}: {fruta}")

# Puedes cambiar el inicio del contador
for pos, fruta in enumerate(frutas, 1):
    print(f"#{pos}: {fruta}")
```

```python
# range() — generar secuencias de números
# range(fin), range(inicio, fin), range(inicio, fin, paso)

print("range(5):", list(range(5)))          # 0, 1, 2, 3, 4
print("range(1,6):", list(range(1, 6)))     # 1, 2, 3, 4, 5
print("range(0,10,2):", list(range(0,10,2))) # Pares
print("range(10,0,-1):", list(range(10,0,-1))) # Cuenta regresiva

# Suma de los primeros 100 números
total = 0
for n in range(1, 101):
    total += n
print(f"
Suma de 1 a 100: {total}")
# También: sum(range(1, 101))
```

## 4.2 List Comprehensions ⭐

Esta es una de las características más poderosas y elegantes de Python.

```python
# Sin list comprehension:
cuadrados = []
for x in range(1, 6):
    cuadrados.append(x ** 2)
print("Sin comprehension:", cuadrados)

# CON list comprehension — una sola línea!
cuadrados = [x ** 2 for x in range(1, 6)]
print("Con comprehension:", cuadrados)

# Sintaxis: [expresión for elemento in iterable]
# Equivalente SQL: SELECT expresión FROM iterable

# Más ejemplos:
nombres = ['ana', 'carlos', 'beatriz']
nombres_upper = [n.upper() for n in nombres]
print(nombres_upper)

# Con condición (filtro)
numeros = range(1, 21)
pares = [n for n in numeros if n % 2 == 0]
print("Pares del 1 al 20:", pares)
```

```python
# List comprehensions más avanzadas
# Transformación + filtro simultáneo
palabras = ['python', 'java', 'c++', 'javascript', 'go', 'rust']

# Solo palabras de más de 3 letras, en mayúsculas
largas = [p.upper() for p in palabras if len(p) > 3]
print("Palabras largas:", largas)

# Múltiples for (producto cartesiano)
pares = [(x, y) for x in [1, 2, 3] for y in ['a', 'b']]
print("Pares:", pares)

# Anidado: tabla de multiplicar 1-5
tabla = [[i * j for j in range(1, 6)] for i in range(1, 6)]
for fila in tabla:
    print(fila)
```

## 4.3 Tuplas — Listas Inmutables

```python
# Las tuplas son como listas pero INMUTABLES
# Se definen con paréntesis ()
coordenadas = (10.5, -20.3)
colores_rgb = (255, 128, 0)

print(type(coordenadas))  # <class 'tuple'>
print(coordenadas[0])     # Acceso igual que lista

# NO se puede modificar:
# coordenadas[0] = 5  # ← TypeError!

# Usar tupla cuando los datos no deben cambiar:
# coordenadas geográficas, colores, configuraciones

# Desempaquetado de tuplas
x, y = coordenadas
r, g, b = colores_rgb
print(f"X: {x}, Y: {y}")
print(f"R: {r}, G: {g}, B: {b}")

# Swap de variables elegante (usa tuplas internamente)
a, b = 10, 20
a, b = b, a  # ¡Intercambia sin variable temporal!
print(f"a={a}, b={b}")
```

### Ejercicio 4.1: Análisis con List Comprehensions

```python
# Ejercicio 4.1
# Dada una lista de temperaturas en Celsius
temps_celsius = [0, 10, 20, 25, 30, 35, 37, 38.5, 40]

# 1. Convierte todas a Fahrenheit usando list comprehension
temps_fahrenheit = [round((c * 9/5) + 32, 1) for c in temps_celsius]
print("Celsius:    ", temps_celsius)
print("Fahrenheit: ", temps_fahrenheit)

# 2. Obtén solo las temperaturas de fiebre (>37.5°C)
fiebre = [c for c in temps_celsius if c > 37.5]
print(f"
Temperaturas de fiebre: {fiebre}")

# 3. Crea lista de tuplas (celsius, fahrenheit)
pares = [(c, f) for c, f in zip(temps_celsius, temps_fahrenheit)]
for c, f in pares:
    print(f"  {c}°C = {f}°F")
```

### Ejercicio 4.2: Tabla de Multiplicar

```python
# Ejercicio 4.2: Genera la tabla del 1 al 10 para un número dado
numero = 7

print(f"=== TABLA DEL {numero} ===")
for i in range(1, 11):
    resultado = numero * i
    barra = "█" * (resultado // 5)  # Visualización
    print(f"{numero} × {i:2d} = {resultado:3d}  {barra}")

# Bonus: Genera TODAS las tablas del 1 al 10 con list comprehension
print("
=== TABLA PITAGÓRICA ===")
header = "    " + "".join(f"{i:4d}" for i in range(1, 11))
print(header)
print("    " + "-" * 40)
for i in range(1, 11):
    fila = f"{i:3d}|" + "".join(f"{i*j:4d}" for j in range(1, 11))
    print(fila)
```

## 🚀 Mini Proyecto: Análisis de Calificaciones

```python
# MINI PROYECTO: Sistema de análisis de calificaciones
import statistics

calificaciones = {
    'Ana': [85, 92, 78, 95, 88],
    'Carlos': [70, 75, 80, 72, 78],
    'Beatriz': [95, 98, 92, 96, 100],
    'Diego': [60, 65, 58, 70, 62],
    'Elena': [88, 85, 90, 87, 91]
}

print("=" * 55)
print(f"{'REPORTE DE CALIFICACIONES':^55}")
print("=" * 55)
print(f"{'Estudiante':<12} {'Promedio':>9} {'Min':>6} {'Max':>6} {'Estado':>10}")
print("-" * 55)

for nombre, notas in calificaciones.items():
    promedio = sum(notas) / len(notas)
    minimo = min(notas)
    maximo = max(notas)
    estado = "✅ Aprobado" if promedio >= 70 else "❌ Reprobado"
    print(f"{nombre:<12} {promedio:>9.1f} {minimo:>6} {maximo:>6} {estado:>10}")

# Estadísticas generales
todos_los_promedios = [sum(n)/len(n) for n in calificaciones.values()]
print("-" * 55)
print(f"
📊 Promedio general del grupo: {sum(todos_los_promedios)/len(todos_los_promedios):.1f}")
print(f"🏆 Mejor estudiante: {max(calificaciones, key=lambda k: sum(calificaciones[k])/len(calificaciones[k]))}")
print(f"📉 Necesita apoyo:   {min(calificaciones, key=lambda k: sum(calificaciones[k])/len(calificaciones[k]))}")
```
