## 1.1 ¿Por qué Python?

Python es un lenguaje de propósito general que destaca por:

| Característica | Descripción |
|---|---|
| **Legibilidad** | El código se parece al inglés natural |
| **Versatilidad** | Web, data science, IA, automatización |
| **Ecosistema** | Miles de librerías disponibles |
| **Comunidad** | Una de las más grandes del mundo |

**Comparación rápida:**
```java
// Java
System.out.println("Hola Mundo");

// Python equivalente:
print("Hola Mundo")
```

````python
# Tu primer programa en Python
# La función print() muestra texto en pantalla
print("¡Hola, Python!")
print("Bienvenido al mundo de la programación con Python")
````

## 1.2 El Intérprete Interactivo

Python puede ejecutarse de dos formas:
1. **Interactivo** (línea a línea) — ideal para probar cosas
2. **Scripts** (archivos .py) — para programas completos

En Jupyter ya estás usando un entorno interactivo.

````python
# Python como calculadora — prueba el intérprete
2 + 3
````

````python
# Operaciones básicas
print(10 * 5)
print(2 ** 10)   # 2 elevado a la 10
print(17 / 3)    # División normal
print(17 // 3)   # División entera
print(17 % 3)    # Módulo (resto)
````

## 1.3 Comentarios en Python

Los comentarios son notas para los humanos — Python los ignora.

````python
# Esto es un comentario de una línea
print("Los comentarios no se ejecutan")  # Comentario al final de línea

# En Python NO existe /* comentario multilínea */ como en C/Java
# Se usan múltiples # o strings con triple comilla (como docstrings)
"""
Este es un string multilínea que a veces se usa como comentario.
No es un comentario oficial, pero funciona en la práctica.
"""
print("Después del bloque de texto")
````
