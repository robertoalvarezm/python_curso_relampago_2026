## 1.1 Tu Primer Programa


````python
# La función print() muestra texto en pantalla
print("¡Hola, Python!")
print("Bienvenido al mundo de la programación con Python")
````

## 1.2 Python como Calculadora


````python
# Operaciones aritméticas básicas
print(2 + 3)      # Suma
print(10 * 5)     # Multiplicación
print(2 ** 10)    # Potencia: 2^10
print(17 / 3)     # División (retorna float)
print(17 // 3)    # División entera
print(17 % 3)     # Módulo (resto)
````

## 1.3 Comentarios


````python
# Comentario de una línea

# En Python no existe /* */ como en C/Java
# Se usan múltiples # o strings con triple comilla como convención

'''
Este string multilinea se usa a veces como comentario de bloque,
aunque oficialmente es solo un literal de string.
'''

print("Los comentarios hacen el codigo mas legible")
````

### Ejercicio 1.2: Calculadora Geométrica


````python
# Calcula el área y perímetro de un círculo con radio 7
radio = 7
pi = 3.14159

area = pi * radio ** 2
perimetro = 2 * pi * radio

print(f"Radio: {radio}")
print(f"Área: {area:.2f}")
print(f"Perímetro: {perimetro:.2f}")
````
