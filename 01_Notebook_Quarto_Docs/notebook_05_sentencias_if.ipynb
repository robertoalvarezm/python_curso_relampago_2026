## 5.1 Condicionales Básicos

```python
# Estructura básica: if / elif / else
# Python usa indentación (4 espacios) en lugar de {}

edad = 22

if edad < 13:
    print("Niño/a")
elif edad < 18:
    print("Adolescente")
elif edad < 65:
    print("Adulto/a")
else:
    print("Adulto/a mayor")

# Operadores de comparación
# ==  igual
# !=  diferente
# <   menor que
# >   mayor que
# <=  menor o igual
# >=  mayor o igual
```

```python
# Tests condicionales — retornan True o False
lenguaje = "python"

print(lenguaje == "python")           # True (case-sensitive)
print(lenguaje == "Python")           # False!
print(lenguaje.lower() == "python")   # True (solución)
print(lenguaje != "java")             # True
print(len(lenguaje) > 3)             # True

# Verificar si está en una colección
lenguajes_web = ['javascript', 'python', 'php', 'ruby']
print("python" in lenguajes_web)      # True
print("java" not in lenguajes_web)    # True
```

```python
# Operadores lógicos: and, or, not
# (Equivalen a && || ! en otros lenguajes)
temperatura = 25
humedad = 80

# and — ambas condiciones deben ser True
if temperatura > 20 and humedad < 60:
    print("Clima agradable")
elif temperatura > 20 and humedad >= 60:
    print("Caluroso y húmedo")
else:
    print("Frío")

# or — al menos una condición debe ser True
nota = 55
if nota >= 90 or nota < 0:
    print("Valor extremo")
else:
    print(f"Nota normal: {nota}")

# not — invierte el valor booleano
lista_vacia = []
if not lista_vacia:  # Si la lista está vacía (Falsy)
    print("La lista no tiene elementos")
```

## 5.2 Valores Truthy y Falsy

Python evalúa condiciones de forma intuitiva.

```python
# En Python, estos valores son considerados False (Falsy):
# False, None, 0, 0.0, "", [], {}, ()

# Esto permite escritura más concisa

nombre = ""
if nombre:
    print(f"Hola, {nombre}")
else:
    print("No se proporcionó nombre")  # Se ejecuta esto

lista = [1, 2, 3]
if lista:
    print(f"Hay {len(lista)} elementos")  # Se ejecuta esto

# Operador ternario (expresión condicional)
# valor_si_true if condicion else valor_si_false
edad = 20
estado = "mayor de edad" if edad >= 18 else "menor de edad"
print(f"Con {edad} años, eres {estado}")
```

## 5.3 Condicionales con Listas

```python
# Verificar condiciones en listas
toppings_disponibles = ['pepperoni', 'champiñones', 'aceitunas', 'queso extra']
pedido = ['pepperoni', 'anchoas', 'champiñones']

print("=== VERIFICANDO PEDIDO ===")
for topping in pedido:
    if topping in toppings_disponibles:
        print(f"  ✅ {topping.title()}: Disponible")
    else:
        print(f"  ❌ {topping.title()}: No disponible")

# Lista vacía
if not pedido:
    print("No hiciste un pedido")
else:
    print(f"
Total de ingredientes: {len(pedido)}")
```

### Ejercicio 5.1: Clasificador de Números

```python
# Ejercicio 5.1
def clasificar_numero(n):
    'Clasifica un numero segun sus propiedades.'
    propiedades = []
    
    if n > 0:
        propiedades.append("positivo")
    elif n < 0:
        propiedades.append("negativo")
    else:
        propiedades.append("cero")
    
    if n != 0:
        if n % 2 == 0:
            propiedades.append("par")
        else:
            propiedades.append("impar")
        
        if n % 3 == 0:
            propiedades.append("múltiplo de 3")
    
    return propiedades

# Prueba con varios números
for num in [-9, -2, 0, 1, 6, 17, 100]:
    props = clasificar_numero(num)
    print(f"{num:4d}: {', '.join(props)}")
```

### Ejercicio 5.2: Sistema de Acceso

```python
# Ejercicio 5.2 — Sistema de control de acceso
usuarios_admin = ['admin', 'superuser', 'root']
usuarios_activos = ['admin', 'ana', 'carlos', 'beatriz']
usuarios_baneados = ['hacker', 'spammer']

def verificar_acceso(usuario, seccion):
    '''Verifica si un usuario tiene acceso a una sección.'''
    usuario = usuario.lower()
    
    if usuario in usuarios_baneados:
        return "🚫 ACCESO DENEGADO — Usuario baneado"
    
    if usuario not in usuarios_activos:
        return "❌ ACCESO DENEGADO — Usuario no registrado"
    
    if seccion == "admin" and usuario not in usuarios_admin:
        return "⚠️  ACCESO DENEGADO — Sin permisos de administrador"
    
    return f"✅ ACCESO CONCEDIDO a sección '{seccion}'"

# Pruebas
pruebas = [
    ('admin', 'admin'),
    ('ana', 'usuario'),
    ('ana', 'admin'),
    ('hacker', 'usuario'),
    ('desconocido', 'usuario'),
]

print("=== SISTEMA DE CONTROL DE ACCESO ===")
for usuario, seccion in pruebas:
    resultado = verificar_acceso(usuario, seccion)
    print(f"Usuario '{usuario}' → Sección '{seccion}': {resultado}")
```

## 🚀 Mini Proyecto: Calculadora de IMC

```python
# MINI PROYECTO: Calculadora de Índice de Masa Corporal (IMC)
def calcular_imc(peso_kg, altura_m):
    '''Calcula el IMC y retorna la clasificación.'''
    imc = peso_kg / (altura_m ** 2)
    
    if imc < 18.5:
        clasificacion = "Bajo peso"
        emoji = "⬇️"
        consejo = "Considera aumentar tu ingesta calórica"
    elif imc < 25:
        clasificacion = "Peso normal"
        emoji = "✅"
        consejo = "¡Mantén tus hábitos saludables!"
    elif imc < 30:
        clasificacion = "Sobrepeso"
        emoji = "⚠️"
        consejo = "Considera aumentar la actividad física"
    else:
        clasificacion = "Obesidad"
        emoji = "🔴"
        consejo = "Consulta a un profesional de salud"
    
    return imc, clasificacion, emoji, consejo

# Casos de prueba
personas = [
    ("Ana", 55, 1.65),
    ("Carlos", 80, 1.75),
    ("Beatriz", 90, 1.60),
    ("Diego", 48, 1.80),
]

print("=" * 60)
print(f"{'CALCULADORA DE IMC':^60}")
print("=" * 60)

for nombre, peso, altura in personas:
    imc, clasif, emoji, consejo = calcular_imc(peso, altura)
    print(f"
{emoji} {nombre}")
    print(f"   Peso: {peso} kg | Altura: {altura} m")
    print(f"   IMC: {imc:.1f} — {clasif}")
    print(f"   💡 {consejo}")
```
