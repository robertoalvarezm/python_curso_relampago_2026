## 2.1 Variables — Tipado Dinámico

Tanto Python como R usan **tipado dinámico**: el tipo se asigna automáticamente según el valor. La diferencia principal está en el operador de asignación.

```python
# Python usa '=' para asignar (R usa '<-' o '=')
# R: mensaje <- "Hola R"
mensaje = "Hola Python"
print(mensaje)

# Tipado dinámico: la variable cambia de tipo según el valor asignado
# R hace lo mismo: x <- 42L luego x <- 3.14 es perfectamente válido
x = 42        # x es int
print(f"x = {x}, tipo: {type(x).__name__}")
x = 3.14      # ahora x es float
print(f"x = {x}, tipo: {type(x).__name__}")
x = "texto"   # ahora x es str
print(f"x = {x}, tipo: {type(x).__name__}")
```

```python
# Reglas de nombrado — diferencia importante con R
# R permite puntos en nombres: mi.variable, datos.limpios
# Python NO usa puntos en variables (reservado para atributos de objetos)

nombre_usuario = "Ana"   # ✅ snake_case — convención Python (equivale a nombre.usuario en R)
# nombre.usuario = "Ana" # ❌ ERROR en Python (operación de atributo)

# Python distingue mayúsculas y minúsculas (igual que R)
nombre = "Python"
Nombre = "Java"
print(nombre, Nombre)  # Son variables DIFERENTES

# Asignación múltiple — no existe en R de esta forma
a, b, c = 1, 2, 3
print(f"a={a}, b={b}, c={c}")

# Intercambiar valores sin variable temporal
# R necesita: temp <- a; a <- b; b <- temp
a, b = b, a
print(f"Después del swap: a={a}, b={b}")
```

## 2.2 Strings (Cadenas de Texto)

En Python los strings son **inmutables** (igual que en R). Se definen con comillas simples o dobles.

```python
# Definición básica — igual que R
# R: frase1 <- 'Hola mundo'
frase1 = 'Hola mundo'
frase2 = "Hola mundo"
print(frase1 == frase2)  # True — son idénticos

# Útil cuando el string contiene comillas internas
apostrofe = "It's a great day"
cita = 'Ella dijo: "Hola"'
print(apostrofe)
print(cita)

# Longitud de un string
# R: nchar("Hola") → Python: len("Hola")
palabra = "Python"
print(f"'{palabra}' tiene {len(palabra)} caracteres")
```

```python
# Métodos de strings — Python usa notación de objeto (string.metodo())
# R usa funciones independientes: toupper(x), tolower(x), trimws(x)
nombre = "  ana patricia  "

# R: toupper(nombre)  →  Python: nombre.upper()
print(nombre.title())    # Cada palabra capitalizada (no existe directo en R base)
print(nombre.upper())    # Todo mayúsculas  — R: toupper()
print(nombre.lower())    # Todo minúsculas  — R: tolower()
print(nombre.strip())    # Eliminar espacios — R: trimws()
print(nombre.lstrip())   # Solo espacios izquierda — R: trimws(side='left')
print(nombre.rstrip())   # Solo espacios derecha  — R: trimws(side='right')
```

```python
# Más métodos útiles de strings
texto = "la ciencia de datos con python es increíble"

# Buscar y reemplazar — R: gsub() / sub()
# R: gsub("python", "R", texto)
print(texto.replace("python", "R"))

# Dividir en lista — R: strsplit()
# R: strsplit(texto, " ")[[1]]
palabras = texto.split(" ")
print(f"Palabras: {palabras[:4]}...")  # Primeras 4
print(f"Total palabras: {len(palabras)}")

# Verificar contenido
print(texto.startswith("la"))    # R: startsWith(texto, "la")
print(texto.endswith("ble"))     # R: endsWith(texto, "ble")
print("python" in texto)         # R: grepl("python", texto)
```

```python
# f-strings — forma moderna de interpolar variables
# R: paste0() o glue::glue() — los f-strings son equivalentes a glue
nombre = "María"
edad = 25
ciudad = "Querétaro"

# Python f-string (desde Python 3.6+)
# R equivalente con glue: glue("Me llamo {nombre}, tengo {edad} años")
presentacion = f"Me llamo {nombre}, tengo {edad} años y vivo en {ciudad}."
print(presentacion)

# Operaciones y métodos dentro del f-string
print(f"En 10 años tendré {edad + 10} años")
print(f"Nombre en mayúsculas: {nombre.upper()}")

# Formateo numérico — muy útil en reportes
precio = 15999.5
print(f"Precio: ${precio:,.2f}")   # Miles con coma, 2 decimales
print(f"Porcentaje: {0.1567:.1%}") # Formato porcentaje
print(f"Científico: {0.000314:.2e}") # Notación científica
```

```python
# Indexing y slicing — DIFERENCIA CLAVE con R
# R usa índices desde 1; Python usa índices desde 0
cadena = "DataScience"

# R: substr(cadena, 1, 4) → "Data"
# Python: cadena[0:4]      → "Data"  (inicio:fin, fin NO incluido)
print(cadena[0])      # Primer carácter — R: substr(cadena, 1, 1)
print(cadena[0:4])    # "Data"         — R: substr(cadena, 1, 4)
print(cadena[4:])     # "Science"      — R: substr(cadena, 5, nchar(cadena))
print(cadena[-7:])    # Últimos 7 chars — R: substr(cadena, nchar(cadena)-6, nchar(cadena))
print(cadena[::-1])   # Invertir string  — R: paste(rev(strsplit(cadena,"")[[1]]),collapse="")
```

## ⚠️ Errores Comunes: Strings

```python
# ERROR 1: Concatenar string con número directamente
# R permite: paste("Valor:", 42)   Python es más estricto
try:
    resultado = "El total es: " + 42  # TypeError!
except TypeError as e:
    print(f"❌ TypeError: {e}")
    print("   → Solución 1: usar f-string   → f'El total es: {42}'")
    print("   → Solución 2: convertir a str → 'El total es: ' + str(42)")

# Ambas soluciones funcionan:
print(f"El total es: {42}")
print("El total es: " + str(42))

# ERROR 2: Índice fuera de rango
cadena = "Hola"
try:
    print(cadena[10])  # IndexError: solo hay 4 caracteres
except IndexError as e:
    print(f"\n❌ IndexError: {e}")
    print("   → Recuerda: índices van de 0 a len(cadena)-1")
```

## 2.3 Números: Integers y Floats

```python
# Enteros (int) — sin límite de tamaño en Python!
# R: integer (32-bit), Python: arbitrariamente grande
a = 1_000_000   # Separador de miles (más legible) — R: 1000000L
b = 0xFF        # Hexadecimal
c = 0b1010      # Binario

print(f"Un millón:   {a:,}")
print(f"Hex 0xFF:    {b} (decimal)")
print(f"Bin 0b1010:  {c} (decimal)")

# Operaciones aritméticas
print("\n--- Operaciones ---")
print(f"División real:    10 / 3  = {10 / 3:.4f}")   # R: 10 / 3
print(f"División entera:  10 // 3 = {10 // 3}")       # R: 10 %/% 3
print(f"Módulo (resto):   10 %% 3  = {10 % 3}")       # R: 10 %% 3  (¡igual!)
print(f"Potencia:         2 ** 10  = {2 ** 10}")       # R: 2 ^ 10  (distinto operador!)
```

```python
# Flotantes (float)
# R y Python tienen el mismo problema de precisión flotante (es de hardware, no del lenguaje)
precio = 9.99
tasa = 0.18
total = precio * (1 + tasa)

print(f"Precio: ${precio}")
print(f"IVA: ${precio * tasa:.2f}")
print(f"Total: ${total:.2f}")

# ⚠️ Problema clásico de punto flotante (igual en R!)
# R: 0.1 + 0.2 == 0.3  → FALSE
print(f"\n0.1 + 0.2 = {0.1 + 0.2}")          # No es exactamente 0.3
print(f"¿Igual a 0.3? {0.1 + 0.2 == 0.3}")   # False!

# Soluciones
print(f"Redondeado: {round(0.1 + 0.2, 1)}")  # R: round(0.1 + 0.2, 1)

import math
print(f"math.isclose: {math.isclose(0.1 + 0.2, 0.3)}")  # R: isTRUE(all.equal(0.1+0.2, 0.3))
```

```python
# Conversión entre tipos (casting)
# R: as.integer(), as.numeric(), as.character()
numero_str = "42"
numero_int = int(numero_str)      # R: as.integer("42")
numero_float = float(numero_str)  # R: as.numeric("42")

print(f"str  → '{numero_str}' ({type(numero_str).__name__})")
print(f"int  → {numero_int} ({type(numero_int).__name__})")
print(f"float→ {numero_float} ({type(numero_float).__name__})")

# De número a string
valor = 3.14
texto = str(valor)  # R: as.character(3.14)
print(f"\nfloat '{valor}' → str '{texto}'")

# Conversión que puede fallar
try:
    int("tres")  # No se puede convertir texto no numérico
except ValueError as e:
    print(f"\n❌ ValueError: {e}")
    print("   → R daría NA con una advertencia; Python lanza error")
```

## 2.4 Caso de Uso Real: Limpieza de Datos

En data science, la limpieza de texto es una tarea cotidiana. Aquí un ejemplo realista.

```python
# Simulación: datos de encuesta con formato inconsistente
# (Como cuando recibes un CSV exportado de Google Forms)
registros_crudos = [
    "  CDMX  ",
    "guadalajara",
    "Monterrey ",
    "  QUERÉTARO",
    "san luis potosi",
]

print("--- Datos originales ---")
for r in registros_crudos:
    print(f"  |{r}|")

# Limpieza: strip + title case
# En R con dplyr: mutate(ciudad = str_trim(str_to_title(ciudad)))
print("\n--- Datos limpios ---")
registros_limpios = []
for r in registros_crudos:
    limpio = r.strip().title()
    registros_limpios.append(limpio)
    print(f"  {limpio}")

print(f"\nTotal registros procesados: {len(registros_limpios)}")
```

```python
# Caso real: parsear y validar una columna de precios
# Datos como vienen de un sistema legacy
precios_raw = ["$1,299.99", "$ 450.00", "800", "1.500,00", "N/A", ""]

print("Precio original    → Valor numérico")
print("-" * 40)
for precio in precios_raw:
    try:
        # Limpieza: quitar $, espacios y comas
        limpio = precio.strip().replace("$", "").replace(",", "").replace(" ", "")
        valor = float(limpio)
        print(f"{precio:<20} → ${valor:>10,.2f}")
    except ValueError:
        print(f"{precio:<20} → {'VALOR INVÁLIDO':>10} ⚠️")
```

```python
# Ejercicio 2.2 — Factura con formato profesional
producto = "Laptop"
precio_unitario = 15999.99
cantidad = 2
descuento = 0.10   # 10%
iva = 0.16         # 16%

# Cálculos
subtotal = precio_unitario * cantidad
monto_descuento = subtotal * descuento
precio_con_descuento = subtotal - monto_descuento
monto_iva = precio_con_descuento * iva
total = precio_con_descuento + monto_iva

# Impresión con formato
print("=" * 45)
print(f"{'FACTURA':^45}")
print("=" * 45)
print(f"Producto: {producto}")
print(f"Precio unitario:  ${precio_unitario:>12,.2f}")
print(f"Cantidad:         {cantidad:>13}")
print(f"Subtotal:         ${subtotal:>12,.2f}")
print(f"Descuento (10%): -${monto_descuento:>12,.2f}")
print(f"IVA (16%):        ${monto_iva:>12,.2f}")
print("-" * 45)
print(f"TOTAL:            ${total:>12,.2f}")
print("=" * 45)
print(f"\n💡 Ahorro con descuento: ${monto_descuento:,.2f} ({descuento:.0%})")
```

### Ejercicio 2.4: Validación de Datos de Entrada

```python
# Ejercicio 2.4 — Validar y convertir un dataset simulado
# Escenario: columna de edades que llegó como strings desde un CSV
# En R lo harías con as.integer() que convierte errores en NA
# En Python debemos manejar los errores explícitamente

edades_raw = ["25", "32", " 28 ", "cuarenta", "19", "", "45.5", "31"]

validas = []
invalidas = []

for valor in edades_raw:
    try:
        edad = int(float(valor.strip()))  # strip + float intermedio maneja "45.5" y " 28 "
        if 0 <= edad <= 120:              # Rango razonable
            validas.append(edad)
        else:
            invalidas.append((valor, "fuera de rango"))
    except (ValueError, AttributeError):
        invalidas.append((valor, "no numérico"))

print(f"✅ Edades válidas ({len(validas)}): {validas}")
print(f"❌ Valores inválidos ({len(invalidas)}):")
for val, razon in invalidas:
    print(f"   '{val}' → {razon}")
print(f"\nPromedio de edades válidas: {sum(validas)/len(validas):.1f} años")
```

## 🚀 Mini Proyecto: Conversor de Unidades con Reporte

```python
# MINI PROYECTO: Conversor de Unidades con reporte formateado
# Útil para reportes científicos donde necesitas múltiples unidades

ANCHO = 50
print("=" * ANCHO)
print(f"{'CONVERSOR DE UNIDADES':^{ANCHO}}")
print("=" * ANCHO)

# Temperatura
celsius = 100
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15
print(f"\n{'TEMPERATURA':^{ANCHO}}")
print("-" * ANCHO)
print(f"  {'Celsius:':<15} {celsius:>8.2f} °C")
print(f"  {'Fahrenheit:':<15} {fahrenheit:>8.2f} °F")
print(f"  {'Kelvin:':<15} {kelvin:>8.2f} K")

# Distancia
km = 42.195  # Maratón
print(f"\n{'DISTANCIA (Maratón)':^{ANCHO}}")
print("-" * ANCHO)
print(f"  {'Kilómetros:':<15} {km:>8.3f} km")
print(f"  {'Metros:':<15} {km * 1000:>8.0f} m")
print(f"  {'Millas:':<15} {km * 0.621371:>8.3f} mi")
print(f"  {'Pies:':<15} {km * 1000 * 3.28084:>8.0f} ft")

# Velocidad
kmh = 120
print(f"\n{'VELOCIDAD':^{ANCHO}}")
print("-" * ANCHO)
print(f"  {'km/h:':<15} {kmh:>8.1f} km/h")
print(f"  {'mph:':<15} {kmh * 0.621371:>8.1f} mph")
print(f"  {'m/s:':<15} {kmh / 3.6:>8.2f} m/s")
print("=" * ANCHO)
```
