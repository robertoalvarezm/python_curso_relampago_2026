| Ver primeras filas | `head(df)` | `df.head()` |
| Dimensiones | `dim(df)` | `df.shape` |
| Tipos de columnas | `glimpse(df)` | `df.info()` |
| Estadísticas | `summary(df)` | `df.describe()` |
| Seleccionar cols | `select(df, a, b)` | `df[['a','b']]` |
| Filtrar filas | `filter(df, a > 5)` | `df[df['a'] > 5]` |
| Nueva columna | `mutate(df, c = a*2)` | `df['c'] = df['a']*2` |
| Renombrar | `rename(df, nuevo=viejo)` | `df.rename(columns={'viejo':'nuevo'})` |
| Ordenar | `arrange(df, desc(a))` | `df.sort_values('a', ascending=False)` |
| Agrupar+resumir | `group_by(df,g) %>% summarise(m=mean(a))` | `df.groupby('g')['a'].mean()` |
| Unir tablas | `left_join(df1, df2, by='id')` | `df1.merge(df2, on='id', how='left')` |

| Valores NA | `is.na(df)` / `drop_na(df)` | `df.isna()` / `df.dropna()` |
| Rellenar NA | `replace_na(df, list(a=0))` | `df.fillna({'a':0})` |
| Aplicar función | `mutate(df, b=map_dbl(a, f))` | `df['b'] = df['a'].apply(f)` |
| Strings | `stringr::str_detect(s, p)` | `df['s'].str.contains(p)` |
| Fechas | `lubridate::year(d)` | `df['d'].dt.year` |
| Gráfica de líneas | `ggplot(df, aes(x,y)) + geom_line()` | `px.line(df, x='x', y='y')` |
| Gráfica de barras | `ggplot(df, aes(x,y)) + geom_col()` | `px.bar(df, x='x', y='y')` |
| Dispersión | `ggplot(df, aes(x,y)) + geom_point()` | `px.scatter(df, x='x', y='y')` |

## 1.1 Series — El Vector Nombrado de Python

````python
# Series: array unidimensional con índice etiquetado
# Equivale a un vector nombrado de R: c(a=1, b=2, c=3)

# Crear desde lista
precios = pd.Series([15999, 349, 799, 5999, 1299],
                    index=['Laptop','Mouse','Teclado','Monitor','Audífonos'],
                    name='precio_mxn')
print(precios)
print(f"\nTipo: {type(precios)}")
print(f"dtype: {precios.dtype}")
print(f"Shape: {precios.shape}")
````

````python
# Acceso — similar a vector nombrado de R
print(precios['Laptop'])          # por etiqueta
print(precios[0])                 # por posición
print(precios[['Laptop','Mouse']]) # múltiples

# Operaciones vectorizadas — como en R, sin loops
con_iva = precios * 1.16
print("\nCon IVA:")
print(con_iva.round(0))

# Filtro booleano
caros = precios[precios > 3000]
print("\nProductos > $3,000:")
print(caros)
````

````python
# Métodos estadísticos — equivalentes directos de R
print(f"min():    {precios.min():,}")
print(f"max():    {precios.max():,}")
print(f"mean():   {precios.mean():,.1f}")
print(f"median(): {precios.median():,}")
print(f"std():    {precios.std():,.1f}")
print(f"sum():    {precios.sum():,}")
print()
print(precios.describe())
````

## 1.5 Selección con `[]`, `.loc` e `.iloc`

````python
# Selección de columnas
print("── Una columna (Series) ───")
print(df_ventas['ventas'])          # con []
print(df_ventas.ventas)             # con .atributo (solo si nombre válido)

print("\n── Múltiples columnas (DataFrame) ───")
print(df_ventas[['vendedor', 'ventas', 'meta']])
````

````python
# .loc — selección por ETIQUETAS (nombres de índice y columna)
# .iloc — selección por POSICIÓN (enteros)

df = df_ventas.copy()
df.index = ['r0','r1','r2','r3','r4','r5','r6','r7']  # índice con etiquetas

print("── .loc por etiqueta ───")
print(df.loc['r0', 'ventas'])                   # celda específica
print(df.loc['r0':'r2', ['vendedor','ventas']])  # rango de filas, cols específicas
print(df.loc[df['region']=='Norte', 'ventas'])   # condición + col

print("\n── .iloc por posición ───")
print(df.iloc[0, 3])                # fila 0, columna 3
print(df.iloc[:3, :3])              # primeras 3 filas, primeras 3 cols
print(df.iloc[-2:, -2:])            # últimas 2 filas, últimas 2 cols
````

## 1.6 Filtrado Booleano

````python
# Filtrado — equivalente de dplyr::filter()

# Condición simple
print("── Ventas > 50,000 ───")
print(df_ventas[df_ventas['ventas'] > 50000])

# Condiciones múltiples — IMPORTANTE: usar & | ~ (no and/or/not)
print("\n── Norte Y ventas > 45,000 ───")
mask = (df_ventas['region'] == 'Norte') & (df_ventas['ventas'] > 45000)
print(df_ventas[mask])

# .query() — sintaxis más legible (como filter() de dplyr)
print("\n── Con .query() ───")
print(df_ventas.query("region == 'Norte' and ventas > 45000"))

# isin() — equivalente de %in% de R
print("\n── Vendedores Ana o Sofía ───")
print(df_ventas[df_ventas['vendedor'].isin(['Ana', 'Sofía'])])

# between() — equivalente de between() de dplyr
print("\n── Ventas entre 40,000 y 55,000 ───")
print(df_ventas[df_ventas['ventas'].between(40000, 55000)])
````

## 1.7 Nuevas Columnas y Transformaciones

````python
# Crear y modificar columnas — equivalente de dplyr::mutate()

df = df_ventas.copy()

# Nueva columna calculada
df['cumplimiento'] = (df['ventas'] / df['meta'] * 100).round(1)
df['precio_unit']  = (df['ventas'] / df['unidades']).round(0)

# Columna condicional con np.where — equivalente de ifelse() de R
df['estado'] = np.where(df['ventas'] >= df['meta'], '✅ Meta', '❌ Bajo meta')

# np.select para múltiples condiciones — equivalente de case_when() de dplyr
condiciones = [
    df['cumplimiento'] >= 110,
    df['cumplimiento'] >= 100,
    df['cumplimiento'] >= 85,
]
categorias = ['⭐ Excelente', '✅ En meta', '⚠️  Cerca']
df['categoria'] = np.select(condiciones, categorias, default='❌ Bajo')

print(df[['vendedor','mes','ventas','meta','cumplimiento','estado','categoria']])
````

````python
# Renombrar, reordenar y eliminar columnas
df2 = df_ventas.copy()

# rename() — equivalente de rename() de dplyr
df2 = df2.rename(columns={'ventas': 'monto', 'unidades': 'qty'})
print("Renombrado:", df2.columns.tolist())

# drop() — eliminar columnas
df2 = df2.drop(columns=['meta'])
print("Sin meta:", df2.columns.tolist())

# Reordenar columnas
nuevo_orden = ['vendedor', 'mes', 'region', 'monto', 'qty']
df2 = df2[nuevo_orden]
print("Reordenado:", df2.columns.tolist())

# astype() — cambiar tipo de datos — equivalente de as.*() de R
df2['monto'] = df2['monto'].astype(float)
df2['mes']   = df2['mes'].astype('category')   # como factor en R
print(f"\nTipos: monto={df2['monto'].dtype}, mes={df2['mes'].dtype}")
````

## 1.8 Valores Faltantes (NaN)

````python
# Crear DataFrame con valores faltantes para demostración
df_na = pd.DataFrame({
    'nombre':  ['Ana', 'Carlos', None, 'Diego', 'Elena'],
    'edad':    [28, None, 25, 34, None],
    'ventas':  [45000, 32000, 58000, None, 41000],
    'region':  ['Norte', 'Sur', 'Norte', None, 'Centro'],
})
print("DataFrame con NaN:")
print(df_na)

# Detectar NA — equivalente de is.na() de R
print("\n── isna() ───")
print(df_na.isna())

print("\n── Conteo de NA por columna ───")
print(df_na.isna().sum())

print("\n── % de NA por columna ───")
print((df_na.isna().sum() / len(df_na) * 100).round(1))
````

````python
# Manejo de valores faltantes
df_na = pd.DataFrame({
    'nombre':  ['Ana', 'Carlos', None, 'Diego', 'Elena'],
    'edad':    [28, None, 25, 34, None],
    'ventas':  [45000, 32000, 58000, None, 41000],
})

# dropna() — equivalente de drop_na() de tidyr
print("── dropna() — eliminar filas con cualquier NA ───")
print(df_na.dropna())

print("\n── dropna(subset) — solo si NA en columna específica ───")
print(df_na.dropna(subset=['ventas']))

# fillna() — equivalente de replace_na() de tidyr
print("\n── fillna() — rellenar NA ───")
df_relleno = df_na.fillna({'edad': df_na['edad'].median(), 'ventas': 0})
print(df_relleno)

# ffill / bfill — relleno hacia adelante/atrás — útil para series de tiempo
df_temporal = pd.DataFrame({'valor': [1, None, None, 4, None, 6]})
df_temporal['ffill'] = df_temporal['valor'].ffill()
df_temporal['bfill'] = df_temporal['valor'].bfill()
print("\n── ffill y bfill ───")
print(df_temporal)
````

## 2.1 `groupby` + `agg` — Equivalente de `group_by` + `summarise`

````python
# groupby — equivalente de group_by() de dplyr

# Una columna, una operación
print("── Ventas totales por vendedor ───")
print(df_ventas.groupby('vendedor')['ventas'].sum())

# Una columna, múltiples operaciones con .agg()
print("\n── Resumen por vendedor ───")
resumen = df_ventas.groupby('vendedor')['ventas'].agg(
    total    = 'sum',
    promedio = 'mean',
    minimo   = 'min',
    maximo   = 'max',
    n_meses  = 'count'
).round(0)
print(resumen)

# Múltiples columnas y operaciones
print("\n── Resumen multi-columna por región ───")
resumen_multi = df_ventas.groupby('region').agg(
    ventas_total  = ('ventas', 'sum'),
    ventas_prom   = ('ventas', 'mean'),
    unidades_tot  = ('unidades', 'sum'),
    n_registros   = ('vendedor', 'count')
).round(0)
print(resumen_multi)
````

````python
# Agrupar por múltiples columnas
print("── Ventas por vendedor y mes ───")
tabla = df_ventas.groupby(['vendedor', 'mes'])['ventas'].sum().reset_index()
print(tabla)

# transform() — como mutate() con group_by en dplyr
# Agrega la estadística del grupo de vuelta al DataFrame original (misma longitud)
df2 = df_ventas.copy()
df2['media_region']    = df2.groupby('region')['ventas'].transform('mean').round(0)
df2['pct_de_region']   = (df2['ventas'] / df2.groupby('region')['ventas'].transform('sum') * 100).round(1)
print("\n── Con transform() ───")
print(df2[['vendedor','region','mes','ventas','media_region','pct_de_region']])
````

::: {.callout-note title="R vs Python — group_by + summarise vs groupby + agg"}
```r
# En R — dplyr
df %>%
  group_by(vendedor) %>%
  summarise(
    total    = sum(ventas),
    promedio = mean(ventas),
    n        = n()
  )

# mutate con group_by (conserva filas)
df %>%
  group_by(region) %>%
  mutate(media_region = mean(ventas))
```
```python
# En Python — pandas
df.groupby('vendedor')['ventas'].agg(
    total    = 'sum',
    promedio = 'mean',
    n        = 'count'
)

# transform (conserva filas, como mutate+group_by)
df['media_region'] = df.groupby('region')['ventas'].transform('mean')
```
El patrón es muy similar. La diferencia principal: pandas usa `agg()` en lugar de `summarise()`, y los parámetros son `nombre=(columna, función)` en lugar de `nombre=función(col)`.

## 2.2 Ordenar y Seleccionar Tops

````python
# sort_values() — equivalente de arrange() de dplyr
print("── Por ventas descendente ───")
print(df_ventas.sort_values('ventas', ascending=False))

# Múltiples columnas
print("\n── Por región ascendente, ventas descendente ───")
print(df_ventas.sort_values(['region', 'ventas'], ascending=[True, False]))

# nlargest / nsmallest — equivalente de top_n() de dplyr
print("\n── Top 3 ventas ───")
print(df_ventas.nlargest(3, 'ventas')[['vendedor','mes','ventas']])

# rank() — equivalente de rank() de R
df2 = df_ventas.copy()
df2['ranking'] = df2['ventas'].rank(ascending=False, method='dense').astype(int)
print("\n── Con ranking ───")
print(df2[['vendedor','mes','ventas','ranking']].sort_values('ranking'))
````

## 2.3 `merge` y `join` — Equivalente de `*_join` de dplyr

````python
# Tablas para demostrar joins
df_vendedores = pd.DataFrame({
    'vendedor':  ['Ana', 'Carlos', 'Sofía', 'Diego', 'Elena'],
    'antigüedad': [5, 3, 2, 7, 1],
    'nivel':     ['Senior', 'Mid', 'Junior', 'Senior', 'Junior'],
})

df_regiones = pd.DataFrame({
    'region':   ['Norte', 'Sur', 'Centro'],
    'director': ['López', 'García', 'Martínez'],
    'meta_regional': [120000, 80000, 90000],
})

# LEFT JOIN — equivalente de left_join() de dplyr
print("── left_join (left merge) ───")
df_left = df_ventas.merge(df_vendedores, on='vendedor', how='left')
print(df_left[['vendedor','mes','ventas','nivel','antigüedad']])

# INNER JOIN
print("\n── inner_join ───")
df_inner = df_ventas.merge(df_regiones, on='region', how='inner')
print(df_inner[['vendedor','region','ventas','director']].head())
````

````python
# Resumen de tipos de join
df_a = pd.DataFrame({'id': [1, 2, 3, 4], 'val_a': ['A1','A2','A3','A4']})
df_b = pd.DataFrame({'id': [2, 3, 4, 5], 'val_b': ['B2','B3','B4','B5']})

tipos = [
    ('inner', 'inner_join — solo coincidencias en AMBOS'),
    ('left',  'left_join  — todos de la izquierda'),
    ('right', 'right_join — todos de la derecha'),
    ('outer', 'full_join  — todos de ambos'),
]

for how, descripcion in tipos:
    resultado = df_a.merge(df_b, on='id', how=how)
    print(f"\n── {descripcion} ───")
    print(resultado)
````

## 2.4 `pivot_table` y `melt` — Equivalente de `pivot_wider`/`pivot_longer`

````python
# pivot_table() — de largo a ancho — equivalente de pivot_wider() de tidyr

print("── Datos originales (formato largo) ───")
print(df_ventas[['vendedor','mes','ventas']])

print("\n── pivot_table (formato ancho) ───")
tabla_ancha = df_ventas.pivot_table(
    values  = 'ventas',
    index   = 'vendedor',
    columns = 'mes',
    aggfunc = 'sum'
)
print(tabla_ancha)

# Con múltiples métricas
print("\n── pivot_table multi-métrica ───")
tabla_multi = df_ventas.pivot_table(
    values  = ['ventas', 'unidades'],
    index   = 'vendedor',
    columns = 'mes',
    aggfunc = 'sum',
    fill_value = 0
)
print(tabla_multi)
````

````python
# melt() — de ancho a largo — equivalente de pivot_longer() de tidyr

df_ancho = pd.DataFrame({
    'vendedor': ['Ana', 'Carlos', 'Sofía'],
    'Ene':      [45000, 32000, 58000],
    'Feb':      [51000, 38000, 62000],
    'Mar':      [53000, 44000, 71000],
})
print("── Datos anchos ───")
print(df_ancho)

df_largo = df_ancho.melt(
    id_vars    = 'vendedor',   # columnas que se mantienen fijas
    value_vars = ['Ene','Feb','Mar'],  # columnas a "derretir"
    var_name   = 'mes',        # nombre de la nueva columna de categorías
    value_name = 'ventas'      # nombre de la columna de valores
)
print("\n── Datos largos (después de melt) ───")
print(df_largo.sort_values(['vendedor','mes']))
````

## 2.5 `apply`, `map` y Operaciones Vectorizadas

````python
# Preferir operaciones vectorizadas sobre apply() cuando sea posible
# apply() es más lento — usarlo solo cuando no hay alternativa vectorizada

df = df_ventas.copy()

# ✅ Vectorizado — siempre más rápido
df['ventas_miles'] = df['ventas'] / 1000
df['cumplimiento'] = df['ventas'] / df['meta']

# apply() a una columna (Series) — para funciones personalizadas
def clasificar_venta(v):
    if v >= 60000: return 'Excelente'
    if v >= 45000: return 'Buena'
    if v >= 30000: return 'Regular'
    return 'Baja'

df['clasificacion'] = df['ventas'].apply(clasificar_venta)

# Equivalente con map() — más conciso para mapeo simple
categorias = {True: '✅ Meta', False: '❌ Bajo'}
df['logro_meta'] = (df['ventas'] >= df['meta']).map(categorias)

print(df[['vendedor','mes','ventas','clasificacion','logro_meta']])
````

````python
# apply() por filas (axis=1) — cuando la función necesita múltiples columnas
# Equivalente de rowwise() + mutate() de dplyr

df = df_ventas.copy()

def score_vendedor(fila):
    """Score compuesto por fila — necesita múltiples columnas."""
    cumplimiento = fila['ventas'] / fila['meta']
    eficiencia   = fila['ventas'] / fila['unidades']
    base_score   = cumplimiento * 50 + min(eficiencia / 5000, 1) * 50
    return round(base_score, 1)

df['score'] = df.apply(score_vendedor, axis=1)   # axis=1 = por filas

print(df[['vendedor','mes','ventas','unidades','meta','score']])
````

    lambda x: x.fillna(x.median())

# d) monto_neto
df['monto_neto'] = (df['monto'] * (1 - df['descuento']) * df['unidades']).round(2)

# e) Separar devoluciones
df_dev  = df[df['devolucion'] == True].copy()
df_ok   = df[df['devolucion'] == False].copy()
print(f"\nDevoluciones: {len(df_dev)} | Ventas válidas: {len(df_ok)}")

# ── 2. Análisis ────────────────────────────────────────────
# a) Por categoría
print("\na) Por categoría:")

# b) Tendencia mensual
df_ok['mes'] = df_ok['fecha'].dt.month
print("\nb) Por mes:")
print(df_ok.groupby('mes')['monto_neto'].sum().round(0))

# c) Por canal
print("\nc) Por canal:")
df_ok['ticket'] = df_ok['monto_neto'] / df_ok['unidades']
print(df_ok.groupby('canal').agg(
    total=('monto_neto','sum'), ticket_prom=('ticket','mean')
).round(0))

# d) Top 10 días
print("\nd) Top 10 días:")
print(df_ok.groupby('fecha')['monto_neto'].sum().nlargest(10).round(0))

# ── 3. Visualización ──────────────────────────────────────
# a) Serie diaria con media móvil
diario = df_ok.groupby('fecha')['monto_neto'].sum().reset_index()
diario['ma7'] = diario['monto_neto'].rolling(7).mean()

fig_a = go.Figure()

## 🚀 Mini Proyecto: Análisis de Resultados Académicos

````python
# Mini Proyecto: Pipeline completo de análisis académico
# Limpieza → Transformación → Análisis → Visualización

np.random.seed(123)
n_alum = 120
materias = ['Matemáticas','Estadística','Programación','Bases de Datos','Machine Learning']

# Generar datos realistas
df_acad = pd.DataFrame({
    'id_alumno':  range(1, n_alum+1),
    'nombre':     [f'Alumno_{i:03d}' for i in range(1, n_alum+1)],
    'semestre':   np.random.choice([3,5,7], n_alum, p=[0.4,0.35,0.25]),
    'modalidad':  np.random.choice(['Presencial','Online'], n_alum, p=[0.6,0.4]),
})
for mat in materias:
    notas = np.random.normal(75, 12, n_alum).clip(0, 100).round(1)
    notas[np.random.choice(n_alum, 8)] = np.nan  # algunos ausentes
    df_acad[mat] = notas

# ── PASO 1: LIMPIEZA ────────────────────────────────────────────────────────
print("=" * 60)
print("PASO 1 — LIMPIEZA")
print("=" * 60)

print(f"NAs por columna:\n{df_acad.isna().sum()[df_acad.isna().sum() > 0]}")

# Imputar notas faltantes con la media de la materia por semestre
for mat in materias:
    df_acad[mat] = df_acad.groupby('semestre')[mat].transform(
        lambda x: x.fillna(x.mean())
    )

assert df_acad[materias].isna().sum().sum() == 0, "Aún hay NaNs"
print("\n✅ Todos los NaN imputados")

# ── PASO 2: TRANSFORMACIÓN ──────────────────────────────────────────────────
print("\n" + "=" * 60)
print("PASO 2 — TRANSFORMACIÓN")
print("=" * 60)

df_acad['promedio']  = df_acad[materias].mean(axis=1).round(1)
df_acad['desv_std']  = df_acad[materias].std(axis=1).round(1)
df_acad['mat_fuerte'] = df_acad[materias].idxmax(axis=1)
df_acad['mat_debil']  = df_acad[materias].idxmin(axis=1)

conds  = [df_acad['promedio']>=90, df_acad['promedio']>=80,
          df_acad['promedio']>=70, df_acad['promedio']>=60]
nivels = ['A','B','C','D']
df_acad['nivel'] = np.select(conds, nivels, default='F')
df_acad['aprobado'] = df_acad['promedio'] >= 70

print(df_acad[['nombre','semestre','modalidad','promedio','nivel','aprobado']].head(8))

# ── PASO 3: ANÁLISIS ────────────────────────────────────────────────────────
print("\n" + "=" * 60)
print("PASO 3 — ANÁLISIS")
print("=" * 60)

# Por semestre
print("\nRendimiento por semestre:")
print(df_acad.groupby('semestre').agg(
    n            = ('promedio','count'),
    prom_general = ('promedio','mean'),
    pct_aprobado = ('aprobado', lambda x: x.mean()*100)
).round(1))

# Por modalidad
print("\nRendimiento por modalidad:")
print(df_acad.groupby('modalidad').agg(
    prom_general = ('promedio','mean'),
    pct_aprobado = ('aprobado', lambda x: x.mean()*100)
).round(1))

# Estadísticas por materia
print("\nEstadísticas por materia:")
df_melt = df_acad.melt(
    id_vars=['semestre','modalidad'],
    value_vars=materias,
    var_name='materia', value_name='nota'
)
print(df_melt.groupby('materia')['nota'].agg(['mean','std','min','max']).round(1))

# Tasa de reprobación por materia
print("\nTasa de reprobación por materia:")
reprobacion = {mat: (df_acad[mat] < 70).mean() * 100 for mat in materias}
for mat, pct in sorted(reprobacion.items(), key=lambda x: -x[1]):
    barra = '█' * int(pct/2)
    print(f"  {mat:<20}: {pct:5.1f}% {barra}")

# ── PASO 4: VISUALIZACIÓN ───────────────────────────────────────────────────
print("\n" + "=" * 60)
print("PASO 4 — VISUALIZACIÓN")
print("=" * 60)

# Matplotlib: Distribución de promedios y barras de tasa
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Histograma por modalidad
for mod, grp in df_acad.groupby('modalidad'):
    axes[0].hist(grp['promedio'], bins=15, alpha=0.6, label=mod,
                 color='steelblue' if mod=='Presencial' else 'coral')
axes[0].axvline(70, color='red', linestyle='--', linewidth=1.5, label='Mínimo aprobatorio')
axes[0].set_title('Distribución de Promedios por Modalidad', fontweight='bold')
axes[0].set_xlabel('Promedio')
axes[0].set_ylabel('Frecuencia')
axes[0].legend()
axes[0].spines[['top','right']].set_visible(False)

# Barras de tasa de reprobación
rep_series = pd.Series(reprobacion).sort_values(ascending=True)
colores_rep = ['#e74c3c' if v > 20 else '#f39c12' if v > 10 else '#2ecc71'
               for v in rep_series.values]
axes[1].barh(rep_series.index, rep_series.values, color=colores_rep)
axes[1].set_xlabel('% Reprobación')
axes[1].set_title('Tasa de Reprobación por Materia', fontweight='bold')
axes[1].axvline(rep_series.mean(), color='gray', linestyle='--', label='Promedio')
for i, v in enumerate(rep_series.values):
    axes[1].text(v+0.3, i, f'{v:.1f}%', va='center', fontsize=9)
axes[1].legend()
axes[1].spines[['top','right']].set_visible(False)

plt.tight_layout()
plt.savefig('/tmp/dashboard_academico_mpl.png', dpi=120, bbox_inches='tight')
plt.show()

# Plotly: dashboard interactivo
fig_px = make_subplots(
    rows=2, cols=2,
    subplot_titles=[
        'Boxplot por Materia y Modalidad',
        'Promedio por Semestre',
        'Distribución de Niveles',
        'Correlación entre Materias'
    ]
)

# Boxplot por materia
for i, mat in enumerate(materias[:3]):
    for j, mod in enumerate(df_acad['modalidad'].unique()):
        vals = df_acad[df_acad['modalidad']==mod][mat]
        fig_px.add_trace(
            go.Box(y=vals, name=f'{mat[:8]}/{mod[:1]}',
                   marker_color=['#3498db','#e74c3c'][j],
                   showlegend=(i==0)),
            row=1, col=1
        )

# Promedio por semestre y modalidad
for mod in df_acad['modalidad'].unique():
    d = df_acad[df_acad['modalidad']==mod].groupby('semestre')['promedio'].mean().round(1)
    fig_px.add_trace(
        go.Bar(x=d.index.astype(str), y=d.values, name=mod, showlegend=True),
        row=1, col=2
    )

# Distribución de niveles
nivel_counts = df_acad['nivel'].value_counts()
fig_px.add_trace(
    go.Bar(x=nivel_counts.index, y=nivel_counts.values,
           marker_color=['#2ecc71','#3498db','#f39c12','#e74c3c','#9b59b6'],
           showlegend=False),
    row=2, col=1
)

# Correlación
corr_mat = df_acad[materias].corr()
fig_px.add_trace(
    go.Heatmap(
        z=corr_mat.values,
        x=[m[:8] for m in corr_mat.columns],
        y=[m[:8] for m in corr_mat.index],
        colorscale='RdBu', zmid=0,
        text=corr_mat.round(2).values,
        texttemplate='%{text}',
        showscale=False
    ),
    row=2, col=2
)

fig_px.update_layout(
    height=700,
    title_text='Dashboard Académico — Análisis Completo',
    template='plotly_white',
    barmode='group'
)
fig_px.show()

print("\n✅ Análisis completo. Figuras guardadas.")
````
